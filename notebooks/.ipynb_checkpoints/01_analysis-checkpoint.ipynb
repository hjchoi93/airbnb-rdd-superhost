## Regression Discontinuity Analysis

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

listings_df = pd.read_csv("listings.csv")
calendar_df = pd.read_csv("calendar.csv")
```

```python
import features, rdd, viz
```

```python
filtered_df = features.occupancy_next_30d(listings_df, calendar_df)
```

```python
m = rdd.fit_rdd(filtered_df, "review_scores_rating", 4.8, 0.1, "occupancy_next_30d", None)
```

```python
# plot multiple bandwidths and with cutoff rating of 4.8
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

c = 4.8

res1 = viz.RDD_plot(filtered_df, running_var="review_scores_rating", c=c, h=0.05, y_col="occupancy_next_30d", ax=axes[0], title="RDD (h=0.05)")
res2 = viz.RDD_plot(filtered_df, running_var="review_scores_rating", c=c, h=0.10, y_col="occupancy_next_30d", ax=axes[1], title="RDD (h=0.10)")
res3 = viz.RDD_plot(filtered_df, running_var="review_scores_rating", c=c, h=0.2, y_col="occupancy_next_30d", ax=axes[2], title="RDD (h=0.2)")

fig.suptitle("RDD: Occupancy vs Rating Cutoff (c=4.8) Across Bandwidths", fontsize=14)

fig.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.savefig("/Users/hojoonchoi/Desktop/airbnb-rdd-superhost/reports/figures/rdd_outcome_bandwidths_4.8.png", dpi=200, bbox_inches="tight")
```

```python
# confidence intervals for above results 

print(res1["ci"])
print(res2["ci"])
print(res3["ci"])
```

```python
# confidence intervals for above results 

print(res4["ci"])
print(res5["ci"])
print(res6["ci"])
```

#### Run the same regression but with covariate as "minimum nights" instead of "occupancy rate"

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

c = 4.8

filtered_df["min_nights_cap30"] = filtered_df["minimum_nights"].clip(upper=30) # enforce a threshold of 30 days to prevent potential skewing of data

res10 = viz.RDD_plot(filtered_df, running_var="review_scores_rating", c=c, h=0.05, y_col="occupancy_next_30d", ax=axes[0], title="RDD (h=0.05)")
res11 = viz.RDD_plot(filtered_df, running_var="review_scores_rating", c=c, h=0.10, y_col="occupancy_next_30d", ax=axes[1], title="RDD (h=0.10)")
res12 = viz.RDD_plot(filtered_df, running_var="review_scores_rating", c=c, h=0.2, y_col="occupancy_next_30d", ax=axes[2], title="RDD (h=0.2)")

fig.suptitle("RDD: Minimum Nights vs Rating Cutoff (c=4.8) Across Bandwidths", fontsize=14)

fig.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.savefig("/Users/hojoonchoi/Desktop/airbnb-rdd-superhost/reports/figures/rdd_outcome_bandwidths_min_nights.png", dpi=200, bbox_inches="tight")
```
